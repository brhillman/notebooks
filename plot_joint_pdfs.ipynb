# Make plots of joint probability distributions
Simple example adapted from the seaborn website (https://seaborn.pydata.org/examples/joint_kde.html).

First, create some synthetic data to work with. This creates a random correlated bivariate dataset:

```python
import numpy
rs = numpy.random.RandomState(5)
mean = [0, 0]
cov = [(1, .5), (.5, 1)]
x1, x2 = rs.multivariate_normal(mean, cov, 500).T
```

Use seaborn to plot the joint distribution of this dataset. The default is to make a scatter plot, with the marginal PDFs plotted on the edges of the scatter plot:

```python
%matplotlib inline
import seaborn
pl = seaborn.jointplot(x1, x2)
pl.set_axis_labels('x1', 'x2')
```

We can use different methods to plotting the joint pdf too, like a "hexbin" plot:

```python
pl = seaborn.jointplot(x1, x2, kind='hex')
pl.set_axis_labels('x1', 'x2')
```

Or kernel density estimate:

```python
pl = seaborn.jointplot(x1, x2, kind='kde')
pl.set_axis_labels('x1', 'x2')
```

Seaborn is meant to work seamlessly with the Pandas library, so if you're familiar with that, the plotting is a little more intuitive and stuff is labelled automatically:

```python
import pandas
x1 = pandas.Series(x1, name='Name of x1 variable')
x2 = pandas.Series(x2, name='Name of x2 variable')
pl = seaborn.jointplot(x1, x2, kind='hex')
```
